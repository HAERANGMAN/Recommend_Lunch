```python
import pandas as pd
sales_df = pd.read_csv('./Sales_Records.csv')
sales_df.head() #몇개만 보여주기
```

```python
sales_df_1=sales_df.drop(["Ship Date","Order ID","Sales Channel","Total Revenue"], axis=1)
sales_df_1.head()
```

```python
sales_df_1.columns=["지역","국가","배송타입","배송순위","주문날짜","판매개수","단가","원가","총비용","총이윤"]
```

```python
sales_df_2=sales_df_1.sort_values("국가", ascending=True)
sales_df_2
```

```python
sales_df_3=sales_df_2.reset_index(drop=True) #새로운 인덱스를 삭제할지 말지, 변수새로 해줘야 리셋된값을 저장함
sales_df_3
```
